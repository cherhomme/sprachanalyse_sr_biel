### Die Files sollen auf DE-FR aufgesplittet werden

Zu jeder SR-Sitzung wird ein Teil auf frz, ein Teil auf Deutsch gesprochen. Sie müssen wegen der Füllwörter, aber auch wegen der prätrainierten spacy-Bibliotheken separat behandelt werden.

Aufgabe: Für alle Textfiles in meinem raw-Ordner, erstelle Files, die so heissen:
* YYYY-MM_SR_FR
* YYYY-MM_SR_DE

Wichtig: Da die Schlaufe über alle Textfiles lange dauern wird, muss die Aktivität monitored werden. Dafür soll timeit gebraucht werden.

```python
import timeit
from langdetect import detect
import os
from tqdm import tqdm
```

```python
def update_text(line, text_de, text_fr):
    try:
        lang = detect(line)
        if lang == 'de':
            text_de += line + "\n"
        elif lang == 'fr':
            text_fr += line + "\n"
    except:
        pass
    return text_de, text_fr

# Folder containing all .txt files
input_folder = '/Users/j/Documents/GitHub/abschlussarbeit/pv_textfiles'

# Output folder for processed text files
output_folder = '/Users/j/Documents/GitHub/abschlussarbeit/pv_textfiles_de_fr'

# Variable to track the last successfully processed file
last_successful_file = ""

# Loop through all .txt files in the input folder
for file_name in tqdm(os.listdir(input_folder), desc='Processing Files'):
    if file_name.endswith('.txt'):
        # Construct the full file path
        file_path = os.path.join(input_folder, file_name)

        # Skip files until the last successful file
        if last_successful_file and file_name <= last_successful_file:
            continue

        # Extract the last seven digits before '.txt' in the filename
        filename_prefix = os.path.splitext(file_name)[0][-7:]

        text_de = ""
        text_fr = ""

        try:
            with open(file_path, 'r', encoding='utf-8') as file:
                for line in tqdm(file, desc=f'Processing Lines in {file_name}', unit='line', leave=False):
                    text_de, text_fr = update_text(line, text_de, text_fr)
        except Exception as e:
            print(f"Error processing {file_name}: {str(e)}")
            continue  # Skip to the next file

        # Save text_de to a file with "SR-DE" prefix and original file name
        output_file_de = os.path.join(output_folder, f"SR-DE_{filename_prefix}_{file_name}")
        with open(output_file_de, 'w', encoding='utf-8') as file_de:
            file_de.write(text_de)

        # Save text_fr to a file with "SR-FR" prefix and original file name
        output_file_fr = os.path.join(output_folder, f"SR-FR_{filename_prefix}_{file_name}")
        with open(output_file_fr, 'w', encoding='utf-8') as file_fr:
            file_fr.write(text_fr)

        # Update the last successful file
        last_successful_file = file_name
```
